```python
import pandas as pd
import matplotlib.pyplot as plt

un = pd.read_csv('UNSUPERVISED_LR.csv')
sup = pd.read_csv('SUPERVISED.csv')


un_col = ['#1f77b4', '#ff7f0e', '#2ca02c'] 
sup_color = ['darkblue', 'brown', 'darkgreen']




plt.figure(figsize=(12, 6))

for i, model in enumerate(un["Model"].unique()):
    subset = un[un["Model"] == model] 
    sup_subset = sup[sup["Model"] == model]  
    plt.plot(subset["n_Epochs"], subset["ROC_AUC"], marker='o', linestyle='-', 
            color=un_col[i % len(un_col)], linewidth=5, markersize=10, label=model+'_unsupervised')
    plt.plot(sup_subset["n_Epochs"], sup_subset["ROC_AUC"], marker='o', linestyle='-', 
            color=sup_color[i % len(sup_color)], linewidth=5, markersize=10, label=model+'_supervised')
    

# Labels and title
plt.xlabel("Epochs", fontsize=16)
plt.ylabel("ROC-AUC", fontsize=16)
plt.title("ROC-AUC for different Models", fontsize=16)

# Show grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)



plt.legend(title="Model", fontsize=12, loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()

# Save and show the plot
plt.savefig("compare.png")  
plt.show()
```
